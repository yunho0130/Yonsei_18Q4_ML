```python
# Set saved file path
file_folder_path = '/Users/maeng-dev/dev/yonsei/MachineLearning/trading/csv/'
origin_file_name = 'K200A1_1Min.csv'
```

```python
# Data Load 
df = pd.read_csv(file_folder_path+origin_file_name)

# Preview
df.head()
```

```python
# Testing colum

## Testing duplication
print df.Up.equals(df.Down)
print df.Open.equals(df.High)
print df.High.equals(df.Low)
print df.Low.equals(df.Close)
print df.Close.equals(df.Open)

## Print Head
df.where(df.Open!=df.High).head()

## Print Len
len_df = len(df)
print len(df)
```

```python
# Rabeling for One hot encoding 
# Rabel: 'Long', 'Short', 'Hold'

# Just for use testing to decrease calculation time
len_df = 100
df = df.head(len_df)
# df.head()

# Unit of one tick. You should consider transaction fee also. 
unit_tick = 0.05

# Note: if you want to get more clear result,
# you have to use 'Decimal Data Structure' instead of 'floating number'
for i in range(len(df)):
#     print df.Close[i]
#     print df.Close[i-1]
    if i == 0:
        df['Long'] = 0
        df['Short'] = 0
        df['Hold'] = 0
    # Long
    elif df.Close[i]-df.Close[i-1]-unit_tick > 0: 
        df['Long'][i] = 1
        df['Short'][i] = 0
        df['Hold'][i] = 0
        print str(i) +" / "+ str(len_df)
    # Short
    elif df.Close[i]-df.Close[i-1]-unit_tick < 0:
        df['Long'][i] = 0
        df['Short'][i] = 1
        df['Hold'][i] = 0
        print str(i) +" / "+ str(len_df)
    # Hold
    elif df.Close[i]-df.Close[i-1]-unit_tick == 0:
        df['Long'][i] = 0
        df['Short'][i] = 0
        df['Hold'][i] = 1
        print str(i) +" / "+ str(len_df)
df.head()
```

```python
import time
timestr = time.strftime("%Y%m%d-%H,,,%M%S")

df.to_csv(file_folder_path+"Labeled_"+timestr+origin_file_name)
```
